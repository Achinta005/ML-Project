```python
import os
import pandas as pd
import glob
import shutil

SPLIT_DIR = '/content/drive/MyDrive/Dataset/splitted-data/'
LOCAL_DIR = '/content/splitted-data/'

os.makedirs(LOCAL_DIR, exist_ok=True)

drive_files = sorted(glob.glob(SPLIT_DIR + '*.feather'))
print(f"Found {len(drive_files)} feather parts in Drive")

for f in drive_files:
    shutil.copy(f, LOCAL_DIR)
print("All parts copied locally")

files = sorted(glob.glob(LOCAL_DIR + '*.feather'))
dfs = []
for i, f in enumerate(files):
    try:
        chunk = pd.read_feather(f)
        dfs.append(chunk)
    except Exception as e:
        print(f"Error loading {f}: {e}")

merged_df = pd.concat(dfs, ignore_index=True)
print(f"\nFinal merged DataFrame shape: {merged_df.shape}")
```
